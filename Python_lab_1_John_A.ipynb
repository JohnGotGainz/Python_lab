```python
class Farm:
    def __init__(self):
        self.animals = {
            'big': [],
            'medium': [],
            'small': []
        }

    def add_animal(self, category, animal):
        if category in self.animals:
            self.animals[category].append(animal)
            print(f"{animal} added to {category} category.")
        else:
            print(f"Invalid category: {category}")

    def remove_animal(self, category, animal):
        if category in self.animals:
            if animal in self.animals[category]:
                self.animals[category].remove(animal)
                print(f"{animal} removed from {category} category.")
            else:
                print(f"{animal} not found in {category} category.")
        else:
            print(f"Invalid category: {category}")

    def display_animals(self):
        print("Farm Animals:")
        for category, animals in self.animals.items():
            print(f"{category.capitalize()} animals:")
            if animals:
                print(", ".join(animals))
            else:
                print("No animals in this category.")


def main():
    farm = Farm()

    while True:
        print("\nWhat do you want to do?")
        print("1. Add an animal")
        print("2. Remove an animal")
        print("3. Display animals")
        print("4. Exit")
        choice = input("Enter your choice : ")

        if choice == '1':
            category = input("Enter the category (big/medium/small): ")
            animal = input("Enter the animal name: ")
            farm.add_animal(category, animal)
        elif choice == '2':
            category = input("Enter the category (big/medium/small): ")
            animal = input("Enter the animal name: ")
            farm.remove_animal(category, animal)
        elif choice == '3':
            farm.display_animals()
        elif choice == '4':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")


if __name__ == "__main__":
    main()
```

```python
import json

class Farm:
    def __init__(self):
        self.animals = {
            'big': [],
            'medium': [],
            'small': []
        }

    def add_animal(self, category, animal):
        if category in self.animals:
            self.animals[category].append(animal)
            print(f"{animal} added to {category} category.")
        else:
            print(f"Invalid category: {category}")

    def remove_animal(self, category, animal):
        if category in self.animals:
            if animal in self.animals[category]:
                self.animals[category].remove(animal)
                print(f"{animal} removed from {category} category.")
            else:
                print(f"{animal} not found in {category} category.")
        else:
            print(f"Invalid category: {category}")

    def display_animals(self):
        print("Farm Animals:")
        for category, animals in self.animals.items():
            print(f"{category.capitalize()} animals:")
            if animals:
                print(", ".join(animals))
            else:
                print("No animals in this category.")

    def save_to_file(self, filename):
        with open(filename, 'w') as file:
            json.dump(self.animals, file)
        print("Farm data saved to", filename)

    def load_from_file(self, filename):
        with open(filename, 'r') as file:
            self.animals = json.load(file)
        print("Farm data loaded from", filename)


def main():
    farm = Farm()

    while True:
        print("\nWhat do you want to do?")
        print("1. Add an animal")
        print("2. Remove an animal")
        print("3. Display animals")
        print("4. Save farm data to file")
        print("5. Load farm data from file")
        print("6. Exit")
        choice = input("Enter your choice : ")

        if choice == '1':
            category = input("Enter the category (big/medium/small): ")
            animal = input("Enter the animal name: ")
            farm.add_animal(category, animal)
        elif choice == '2':
            category = input("Enter the category (big/medium/small): ")
            animal = input("Enter the animal name: ")
            farm.remove_animal(category, animal)
        elif choice == '3':
            farm.display_animals()
        elif choice == '4':
            filename = input("Enter filename to save: ")
            farm.save_to_file(filename)
        elif choice == '5':
            filename = input("Enter filename to load: ")
            farm.load_from_file(filename)
        elif choice == '6':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")


if __name__ == "__main__":
    main()
```

```python
import random

class Player:
    def __init__(self, name, player_type):
        self.name = name
        self.player_type = player_type
        self.inventory = []
        self.attributes = {
            'health': 100,
            'damage': 10,
            'armor': 5
        }

    def attack(self, other_player):
        damage_dealt = random.randint(1, 10) * self.attributes['damage']
        damage_dealt -= other_player.attributes['armor']
        other_player.attributes['health'] -= damage_dealt
        return damage_dealt

    def display_inventory(self):
        print(f"{self.name}'s Inventory:")
        for item in self.inventory:
            print("-", item)
        print()

    def display_attributes(self):
        print(f"{self.name}'s Attributes:")
        for attr, value in self.attributes.items():
            print(f"{attr.capitalize()}: {value}")
        print()


def main():
    print("Welcome to Player vs Player Game!\n")

    player1_name = input("Enter main characters name: ")
    player1_type = input("Enter main characters type (titian, warlock, hunter, demigod): ")
    player1 = Player(player1_name, player1_type)

    player2_name = input("\nEnter Player 2's name: ")
    player2_type = input("Enter Player 2's type (titian, warlock, hunter, demigod): ")
    player2 = Player(player2_name, player2_type)

    print("\nPlayers created!\n")

    while player1.attributes['health'] > 0 and player2.attributes['health'] > 0:
        print(f"{player1.name} attacks {player2.name}!")
        damage_dealt = player1.attack(player2)
        print(f"{player1.name} dealt {damage_dealt} damage to {player2.name}.")
        print(f"{player2.name}'s health: {player2.attributes['health']}\n")

        if player2.attributes['health'] <= 0:
            print(f"{player1.name} wins!")
            break

        print(f"{player2.name} attacks {player1.name}!")
        damage_dealt = player2.attack(player1)
        print(f"{player2.name} dealt {damage_dealt} damage to {player1.name}.")
        print(f"{player1.name}'s health: {player1.attributes['health']}\n")

        if player1.attributes['health'] <= 0:
            print(f"{player2.name} wins!")
            break

    print("\nGame Over!")


if __name__ == "__main__":
    main()
```
